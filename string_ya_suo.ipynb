```python


# 编码规则为: k[encoding_string], 表示重复k次encoding_strng, 
# 例如'abcdefabcdefabc'可表示为'2[abcdef]abc', 但是'aaa'仅能编码成'aaa', 
# 因为len('3[a]')>len('aaa').
# 补充:
# 1. k为正整数, []内的encoding_string不得含有空格不得为空;
# 2. []内的encoding_string 本身可以为编码过的字符串, 例如'abcdabcdeabcdabcde' 可以编码为 '2[abcdabcde]'(编码后长度从18减少到12), []内的'abcdabcde'又可以编码为 '2[abcd]e', 最终编码为 '2[2[abcd]e]', 编码后长度为11, 应返回11; 这个编码路径也能是: 'abcdabcdeabcdabcde' -> '2[abcd]e2[abcd]e' -> '2[2[abcd]e]';
# 2. 输入字符串为全小写英文字母, 长度<=160;
# 3. 如果编码后长度没有更小, 则保留原有字符串;


def str_pro(s):
    len1 = len(s)   # 字符串长度
    print('len1', len1)
    len2 = len1 // 2   # 初始连续字符长度，最大为字符串长度的一半，由大到小递减
    if len1 <= 4:  # 如果字符串长度小于4，长度不会缩减，直接返回
        return s
    global best_count # 最优循环次数
    best_count=1
    global best_len # 最优缩减后字符长度
    best_len=len1
    print('best_len', best_len)
    while (len2 >= 1): # 循环字符最小长度为1
        for i in range(0, len1 - len2 * 2 + 1): # 循环字符的起始位 相当于先固定循环单元的长度，然后左右移动两倍得循环单元，
        # 然后左右移动找不到使这两个相等的话，就需要改变循环单元的长度，然后接着左右移动。直到发现那个能使两个循环单元前后相等的位置与长度。
            print('i', i)
            count = 1 # 记录循环次数
            s1 = s[i:i + len2] # 循环字符串
            s2 = s[i + len2:i + len2 * 2] # 他后面相同长度字符串
            while (s1 == s2):
                count += 1
                print('len1', len1, 'len2', len2, 'count', count, 'i + len2 * (count + 1)', i + len2 * (count + 1))
                
                if i + len2 * (count + 1) <= len1: # 不能超界
                    s2 = s[i + len2 * count:i + len2 * (count + 1)]
                    print('没越界')
                else:
                    break
            
            newline=len1-len2*count+len(str(count))+2+len2 # 新的字符串长度 '2[]' + 括号里面的字符长度
            print('newline', newline)
            print('best_len', best_len)
            print('count', count)
            if count > 1 and newline < best_len: # 和当前最优比较
                best_len=newline # 更新
                best_count=count
                pre = s[:i] # 将新的字符串分为 前串  循环字符串  后串
                cur = s[i:i + len2]
                lat = s[i + len2 * count:]
                print('pre', pre)
                print('cur', cur)
                print('lat', lat)
        len2 -= 1 # 循环字符串长度减1
        print('best_count', best_count)
        print('len2 -', len2)
    if best_count==1: # 未缩减
        return s
    # 前串  循环字符串  后串 分别递归
    return str_pro(pre) + str(count) + '[' + str_pro(cur) + ']' + str_pro(lat)


if __name__ == '__main__':
    s = strt
    result = str_pro(s)
    print(len(result))
    
    
```
